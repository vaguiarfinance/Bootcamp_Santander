## **E**xtract

Extraia a lista de IDs de usu√°rio a partir do arquivo CSV. Para cada ID, fa√ßa uma requisi√ß√£o GET para obter os dados do usu√°rio correspondente.

````python
import pandas as pd
````

````python
df = pd.read_excel('Santander_Coders.xlsx')
user_ids = df['UserID'].tolist()
print(user_ids)
````

````python
import requests
import json

def get_user(id):
  response = requests.get(f'{sdw2023_api_url}/users/{id}')
  return response.json() if response.status_code == 200 else None

users = [user for id in user_ids if (user := get_user(id)) is not None]
print(json.dumps(users, indent=2))
````

````python
import openai
from PIL import Image, ImageDraw, ImageFont
import textwrap
````

````python
openai.api_key = openai_api_key

def generate_ai_news(user):
  completion = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
      {
          "role": "system",
          "content": "Voc√™ √© um especialista em Mercado Financeiro."
      },
      {
          "role": "user",
          "content": f"Crie uma mensagem para {user['name']} sobre a import√¢ncia dos investimentos (m√°ximo de 100 caracteres)"
      }
    ]
  )
  return completion.choices[0].message.content.strip('\"')

def format_news(news):
    return f"\033[41m\033[1m{news.upper()}\033[0m"

for user in users:
    news = generate_ai_news(user)

    # Formata a not√≠cia gerada em caixa alta e com fundo amarelo
    formatted_news = format_news(news)

    # Imprime a not√≠cia formatada
    print(formatted_news)

    # Adiciona a not√≠cia formatada ao dicion√°rio do usu√°rio
    if 'news' in user:
        user['news'].append({
            "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
            "description": formatted_news
        })
    else:
        user['news'] = [{
            "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
            "description": formatted_news
        }]

````

## **L**oad

Atualize a lista de "news" de cada usu√°rio na API com a nova mensagem gerada.

````python
def update_user(user):
  response = requests.put(f"{sdw2023_api_url}/users/{user['id']}", json=user)
  return True if response.status_code == 200 else False

for user in users:
    success = update_user(user)

    if success:
        print(f"User {user['name']} updated!")
    else:
        print("Usu√°rio n√£o identificado no sistema, favor cadastrar!")
````
